finding how many datatypes will be there.

```python
def classify_feature(col, series):
    if pd.api.types.is_integer_dtype(series):
        if "id" in col.lower():
            return "NUM_INT_ID"
        elif series.nunique() == 2:
            return "NUM_INT_BINARY"
        elif "score" in col.lower():
            return "NUM_INT_RATING"
        elif "status" in col.lower():
            return "NUM_INT_RATING"
        elif "amount" in col.lower():
            return "NUM_INT_QTY"
        elif "term" in col.lower():
            return "NUM_INT_COUNT"
        elif "count" in col.lower() or "loan" in col.lower():
            return "NUM_INT_COUNT"
        else:
            return "NUM_INT_COUNT"

    elif pd.api.types.is_float_dtype(series):
        if "ratio" in col.lower():
            return "NUM_FLOAT_RATIO"
        elif "percent" in col.lower():
            return "NUM_FLOAT_PERCENT"
        else:
            return "NUM_FLOAT_RATIO"  # fallback

    else:
        return "UNKNOWN"

# Apply the classification
feature_types = {}
for col in df.columns:
    feature_types[col] = classify_feature(col, df[col])

# Print result
for k, v in feature_types.items():
    print(f"● {k} = \"{v}\"")
```

```python
def classify_features(series):
    feature_info = []

    for col in df.columns:
        dtype = df[col].dtype
        nunique = df[col].nunique()
        if dtype == 'object':
            feat_type = 'TEXT'
            if nunique == df.shape[0]:
                label_type = 'UNIQUE_ID'
            elif nunique <= 10:
                label_type = 'CATEGORY'
            else:
                label_type = 'LONG_TEXT'
        elif dtype in ['int64', 'float64']:
            if nunique == df.shape[0]:
                feat_type = 'ID/KEY'
                label_type = 'UNIQUE_ID'
            elif nunique <= 2:
                feat_type = 'BINARY'
                label_type = 'BOOLEAN'
            elif dtype == 'int64':
                feat_type = 'COUNT'
                label_type = 'DISCRETE'
            else:
                feat_type = 'RATIOS'
                label_type = 'CONTINUOUS'
        else:
            feat_type = 'UNKNOWN'
            label_type = 'UNKNOWN'

        feature_info.append([col, str(dtype), feat_type, label_type])

    return pd.DataFrame(feature_info, columns=['Column', 'DataType', 'FeatType', 'LabelType'])

# Run it:
classify_features(df)
```

```python
def classify_ranges(series):
    if pd.api.types.is_numeric_dtype(series):
        min = series.min()
        max = series.max()
        if min == 1 and max == 1:
            return "1 -> refine"
        elif 0 <= min and max <= 1:
            return "0–1 - normalize"
        else:
            return f"{min}-{max} -> consider scaling"
    else:
        return "non-numeric -> consider encoding"
        print('value range classification:')
df.apply(classify_ranges)
```
